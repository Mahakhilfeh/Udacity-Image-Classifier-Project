## Create Pipeline

````python
# Create a pipeline for each set.
# Function to create a pipeline for each dataset
def create_pipeline(dataset):
    def process_data(image, label):
        image = tf.image.resize(image, (224, 224))  # Resize the image to 224x224
        image = tf.cast(image, tf.float32) / 255.0  # Normalize pixel values to the range [0, 1]
        return image, label

    return dataset.map(process_data).batch(32).prefetch(tf.data.experimental.AUTOTUNE)  # Apply transformations and optimize data pipeline

# Creating the Pipelines for different datasets
train_pipeline = create_pipeline(train_set)  # Training dataset pipeline
valid_pipeline = create_pipeline(valid_set)  # Validation dataset pipeline
test_pipeline = create_pipeline(test_set)    # Testing dataset pipeline


````

# Build and Train the Classifier

````python
import tensorflow_hub as hub
from tensorflow.keras import models, layers
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# Load the pre-trained model
mobilenet_url = "https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4"
mobilenet_model = hub.KerasLayer(mobilenet_url, input_shape=(224, 224, 3), trainable=False)

# Define the number of classes (flower types)
num_classes = 102

# Build the classification model
model = models.Sequential([
    mobilenet_model,
    layers.Dense(1024, activation='relu'),  # Fully connected layer with 1024 units
    layers.Dropout(0.3),  # Dropout layer to prevent overfitting
    layers.Dense(512, activation='relu'),  # Fully connected layer with 512 units
    layers.Dropout(0.3),  # Dropout layer to prevent overfitting
    layers.Dense(num_classes, activation='softmax')  # Output layer with softmax for classification
])

# Display model summary
model.summary()

# Compile the model
model.compile(optimizer='adam',  # Using Adam optimizer
              loss='sparse_categorical_crossentropy',  # Sparse categorical crossentropy loss
              metrics=['accuracy'])  # Track accuracy

# Early stopping callback to stop training when validation loss is not improving
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Model checkpoint to save the best model based on validation loss
model_checkpoint = ModelCheckpoint('best_model.h5', monitor='val_loss', save_best_only=True)

# Train the model
history = model.fit(
    train_pipeline,  # Training data pipeline
    epochs=15,  # Number of epochs
    validation_data=valid_pipeline,  # Validation data pipeline
    callbacks=[early_stopping, model_checkpoint]  # Callbacks to use during training
)

print("✅ Training Complete. Best model saved as 'best_model.h5'")
````

````python
# TODO: Plot the loss and accuracy values achieved during training for the training and validation set.

# Extract training and validation data
train_loss = history.history['loss']
val_loss = history.history['val_loss']
train_acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

epochs = range(1, len(train_loss) + 1)

plt.figure(figsize=(12, 5))

# 📌 Plot the loss curve
plt.subplot(1, 2, 1)
plt.plot(epochs, train_loss, 'bo-', label='Training Loss')  # Training loss
plt.plot(epochs, val_loss, 'ro-', label='Validation Loss')  # Validation loss
plt.xlabel('Epochs')  # Label for x-axis
plt.ylabel('Loss')  # Label for y-axis
plt.title('Training & Validation Loss')  # Title for the plot
plt.legend()  # Show legend

# 📌 Plot the accuracy curve
plt.subplot(1, 2, 2)
plt.plot(epochs, train_acc, 'bo-', label='Training Accuracy')  # Training accuracy
plt.plot(epochs, val_acc, 'ro-', label='Validation Accuracy')  # Validation accuracy
plt.xlabel('Epochs')  # Label for x-axis
plt.ylabel('Accuracy')  # Label for y-axis
plt.title('Training & Validation Accuracy')  # Title for the plot
plt.legend()  # Show legend

plt.show()  # Display the plot
````

### Image Pre-processing

The `process_image` function should take in an image (in the form of a NumPy array) and return an image in the form of a NumPy array with shape `(224, 224, 3)`.

First, you should convert your image into a TensorFlow Tensor and then resize it to the appropriate size using `tf.image.resize`.

Second, the pixel values of the input images are typically encoded as integers in the range 0-255, but the model expects the pixel values to be floats in the range 0-1. Therefore, you'll also need to normalize the pixel values. 

Finally, convert your image back to a NumPy array using the `.numpy()` method.

````python
# TODO: Create the process_image function
import tensorflow as tf
import numpy as np

def preprocess_image_for_inference(image, target_size=(224, 224)):
    """
    Preprocess the input image for inference:
    - Resize to target size (224, 224)
    - Normalize pixel values to [0, 1]
    """
    # Ensure the image is a TensorFlow tensor
    image_tensor = tf.convert_to_tensor(image)
    
    # Resize and normalize
    image_tensor = tf.image.resize(image_tensor, target_size)
    image_tensor = tf.cast(image_tensor, tf.float32) / 255.0
    
    return image_tensor.numpy()
````

### Inference

Remember, the `predict` function should take an image, a model, and then returns the top $K$ most likely class labels along with the probabilities. The function call should look like: 

```python
probs, classes = predict(image_path, model, top_k)
```

If `top_k=5` the output of the `predict` function should be something like this:

```python
probs, classes = predict(image_path, model, 5)
print(probs)
print(classes)
> [ 0.01558163  0.01541934  0.01452626  0.01443549  0.01407339]
> ['70', '3', '45', '62', '55']
```

Your `predict` function should use `PIL` to load the image from the given `image_path`. You can use the [Image.open](https://pillow.readthedocs.io/en/latest/reference/Image.html#PIL.Image.open) function to load the images. The `Image.open()` function returns an `Image` object. You can convert this `Image` object to a NumPy array by using the `np.asarray()` function.

**Note:** The image returned by the `process_image` function is a NumPy array with shape `(224, 224, 3)` but the model expects the input images to be of shape `(1, 224, 224, 3)`. This extra dimension represents the batch size. We suggest you use the `np.expand_dims()` function to add the extra dimension. 

````python
def predict(image_path, model, top_k=5):
    im = Image.open(image_path)
    image_arr = np.asarray(im)
    
    # Preprocess the image correctly
    processed_image = preprocess_image_for_inference(image_arr, (224, 224))

    # Add batch dimension
    processed_image = np.expand_dims(processed_image, 0)

    # Get predictions
    ps = model.predict(processed_image)

    # Get top K probabilities and class indices
    top_k_indices = np.argsort(ps[0])[-top_k:][::-1]
    top_ps = ps[0][top_k_indices]

    # Convert to 1-based class indices
    top_classes = [str(int(index) + 1) for index in top_k_indices]

    return top_ps, top_classes
````

# Sanity Check

It's always good to check the predictions made by your model to make sure they are correct. To check your predictions we have provided 4 images in the `./test_images/` folder:

* cautleya_spicata.jpg
* hard-leaved_pocket_orchid.jpg
* orange_dahlia.jpg
* wild_pansy.jpg

In the cell below use `matplotlib` to plot the input image alongside the probabilities for the top 5 classes predicted by your model. Plot the probabilities as a bar graph. The plot should look like this:

<img src='assets/inference_example.png' width=600px>

You can convert from the class integer labels to actual flower names using `class_names`. 

````python
def predict_and_plot_image(image_path, model, class_names, top_k=5):
    ps, classes = predict(image_path, model, top_k)

    im = Image.open(image_path)
    image_arr = np.asarray(im)

    # Create the plot
    fig, (ax1, ax2) = plt.subplots(figsize=(6,9), ncols=2)
    ax1.imshow(image_arr, cmap = plt.cm.binary)
    ax1.axis('off')

    # Plot the probabilities as a bar chart
    ax2.barh(np.arange(len(ps)), ps)
    ax2.set_aspect(0.1)
    ax2.set_yticks(np.arange(len(ps)))

    # Convert class indices to flower names
    classes_names_list = []
    for key in classes:
        key = str(int(key))  # Convert to string index
        if key in class_names:  # Check if the string key exists in the dictionary
            classes_names_list.append(class_names[key])
        else:
            classes_names_list.append('Unknown Class')

    # Add a fallback mechanism for missing class names
    ax2.set_yticklabels(classes_names_list, size='small')
    ax2.set_title('Class Probability')
    ax2.set_xlim(0, 1.1)

    plt.tight_layout()

# Example usage for the images
image_paths = [
    './test_images/cautleya_spicata.jpg',
    './test_images/hard-leaved_pocket_orchid.jpg',
    './test_images/orange_dahlia.jpg',
    './test_images/wild_pansy.jpg'
]

for image_path in image_paths:
    predict_and_plot_image(image_path, model, class_names, top_k=5)
````
